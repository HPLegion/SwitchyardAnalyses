```python
import os
import altair as alt
alt.enable_mime_rendering()
import pandas as pd
```

```python
DATA = pd.DataFrame()
models = ["emit_200_flatplates_v2_size", "emit_200_sashasbender_size", "emit_200_cylindricalplates_size"]
for filename in os.listdir():
    if "emit" in filename and "size" in filename:
        frame = pd.read_csv(filename)
        frame["modelname"] = filename[8:-9]
        DATA = DATA.append(frame, ignore_index=True)
```

```python
alt.Chart(DATA.loc[DATA["beamsize"]<=20]).mark_line().encode(
    color="modelname",
    x='beamsize',
    y='x_norm_emittance',
)
```
