```python
import os
import altair as alt
alt.enable_mime_rendering()
import pandas as pd
```

```python
DATA = pd.DataFrame()
for filename in os.listdir():
    if "emit_v7" in filename and not "verticalTilt.csv"in filename:
        frame = pd.read_csv(filename)
        frame["modelname"] = filename[5:-4]
        DATA = DATA.append(frame, ignore_index=True)
```

```python
print(DATA.head(5))
print(DATA.tail(5))
```

```python
alt.Chart(DATA).mark_line().encode(
    column="modelname",
    color='y_offset:N',
    x='x_offset',
    #y='x_emittance_rel',
    y=alt.Y("x_emittance_rel", scale=alt.Scale(domain=(1,1.2), zero=False))
)
```

```python
alt.Chart(DATA).mark_line().encode(
    column="modelname",
    color='y_offset:N',
    x='x_offset',
    #y='y_emittance_rel',
    y=alt.Y("y_emittance_rel", scale=alt.Scale(domain=(1,1.2), zero=False))
)
```
