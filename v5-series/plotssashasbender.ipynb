```python
import os
import altair as alt
alt.enable_mime_rendering()
import pandas as pd
```

```python
DATA = pd.DataFrame()
for filename in os.listdir():
    if filename[:10] == "emit_sasha":
        frame = pd.read_csv(filename)
        frame["modelname"] = filename[5:-4]
        DATA = DATA.append(frame, ignore_index=True)
```

```python
DATA
```

```python
alt.Chart(DATA).mark_line().encode(
    color="modelname",
    column='y_offset:N',
    x='x_offset',
    y='x_emittance_rel',
)
```
