```python
import os
import altair as alt
alt.enable_mime_rendering()
import pandas as pd
```

```python
DATA = pd.DataFrame()
models = ["emit_200_flatplates_v2_size", "emit_200_sashasbender_size", "emit_200_cylindricalplates_size"]
for filename in os.listdir():
    if any(model in filename for model in models):
        frame = pd.read_csv(filename)
        frame["modelname"] = filename[5:-4]
        DATA = DATA.append(frame, ignore_index=True)
```

```python
DATA
```

```python
alt.Chart(DATA.loc[DATA["beamsize"]<=25]).mark_line().encode(
    color="modelname",
    x='beamsize',
    y='x_emittance',
)
```

```python
alt.Chart(DATA.loc[DATA["beamsize"]<=25]).mark_line().encode(
    color="modelname",
    x='beamsize',
    y='y_emittance',
)
```
