```python
import os
import altair as alt
alt.enable_mime_rendering()
import pandas as pd
```

```python
DATA = pd.DataFrame()
models = ["emit_v6_base_no_", "emit_200_cylindricalplates.csv"]
for filename in os.listdir():
    if any(model in filename for model in models):
        frame = pd.read_csv(filename)
        frame["modelname"] = filename[5:-4]
        DATA = DATA.append(frame, ignore_index=True)
```

```python
DATA
```

```python
alt.Chart(DATA).mark_line().encode(
    column="modelname",
    color='y_offset:N',
    x='x_offset',
    #y='x_emittance_rel',
    y=alt.Y("x_emittance_rel", scale=alt.Scale(domain=(1,1.2)))
)
```

```python
alt.Chart(DATA).mark_line().encode(
    column="modelname",
    color='y_offset:N',
    x='x_offset',
    #y='y_emittance_rel',
    y=alt.Y("y_emittance_rel", scale=alt.Scale(domain=(1,1.1)))
)
```
